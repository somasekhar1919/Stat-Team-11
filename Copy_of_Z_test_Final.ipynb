```python
import pandas as pd
import numpy as np
from statsmodels.stats.weightstats import ztest 

df = pd.read_csv("/content/drive/MyDrive/apy.csv")
df = df.dropna(how='any')

databo = (df['State_Name']== 'Karnataka')   # data filter 
data_karnataka = df.loc[databo, 'Production']   
```

```python
alpha=0.05 #since our CI is 95%
Production_Mean = np.mean(data_karnataka)
sigma = np.std(data_karnataka)

print("The Mean of Production               =" ,Production_Mean)
print("The Standard Deviation of Production =" ,sigma)
```

# Production Mean = 40962

#      H0: Production is 40962

#      H1: Production is not equal to 40962

```python
sample_size = 2500
np.random.seed(10)
a = np.random.choice(data_karnataka,sample_size)
a
```

```python
a_mean = np.mean(a)
a_mean
```

```python
n = sample_size
Std_1 = sigma/np.sqrt(n)
Std_1
```

```python
data = Std_1*a+Production_Mean
data
```

```python
ztest_score,p_value = ztest(data,value=a_mean)

print(ztest_score,p_value)
```

```python
if(p_value <  alpha):
  print("Reject Null Hypothesis")
else:
  print("Fail to Reject NUll Hypothesis")
```
